```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib notebook
df1=pd.read_csv('corn_state.csv',usecols=['Year','State','Value'])
df1['Value']=df1['Value'].replace(to_replace=' (D)', value='0').str.replace(',', '').astype(float)
df1=df1.sort_values(by=['State'])
Zones=pd.read_csv('time_zone.csv')
Zones['State']=Zones['State'].replace(to_replace='\([0-9a-zA-Z^\)]+\)', value='',regex=True).apply(lambda x: x.upper())
ZoneA=Zones.where(Zones['Time Zone']=='Central Standard Time (CST)').dropna().merge(df1,how='inner',on='State')
ZoneB=Zones.where(Zones['Time Zone']=='Mountain Standard Time (MST)').dropna().merge(df1,how='inner',on='State')
ZoneC=Zones.where(Zones['Time Zone']=='Pacific Standard Time (PST)').dropna().merge(df1,how='inner',on='State')
ZoneD=Zones.where(Zones['Time Zone']=='Eastern Standard Time (EST)').dropna().merge(df1,how='inner',on='State')
Zone=ZoneA.groupby(['Year'])['Value'].agg({'Central Standard Time':np.sum})
Zone['Mountain Standard Time Corns Production']=ZoneB.groupby(['Year'])['Value'].agg({'Mountain Standard Time':np.sum})['Mountain Standard Time']
Zone['Pacific Standard Time Corns Production']=ZoneC.groupby(['Year'])['Value'].agg({'Pacific Standard Time':np.sum})['Pacific Standard Time']
Zone['Eastern Standard Time Corns Production']=ZoneD.groupby(['Year'])['Value'].agg({'Eastern Standard Time':np.sum})['Eastern Standard Time']
Zone.plot();
plt.title('Kernel Density Estimate of Production of Corns in America Standard Time Zone Wise',y=1.08)
```
